```python
class Char:
    
    Code = (dict(zip(range(30, 87), '"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ')))    
    Inventory_seg_num = 13
    Equip_seg_num = 17
    Skill_seg_num = 3
    
    Temp_values = [('Difficulty', 2), 
                   ('Str', 100000),
                   ('Dex', 100000),
                   ('Int', 100000),
                   ('HP', 1000000000),
                   ('MP', 100000)]
    
    Int_values = {'modifierPoints': (60000, 2),
                  'skillPoints': (10000, 2),
                  'money': (1000000000, 4),
                  'experiencePoints': (1000000000, 4),
                  'masteriesAllowed': (10, 1),
                  'numberOfDeaths': (0, 1),
                  'max': (237, 2) # length of Char.skills_df
                 }
    
    itemtype_df = pd.read_csv("itemclass.csv")
    Itemtype = Enum('Itemtype', itemtype_df.drop('class', axis=1).set_index('category').T.to_dict().items())
    items_df = pd.read_csv("items.csv")
    Items = Enum('Items', items_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    skills_df = pd.read_csv("skills.csv")
    Skills = Enum('Skills', skills_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    prefix_df = pd.concat([pd.read_csv(_f) for _f in ["prefix and suffix.csv",
                                                      "prefix.csv"]])
    Prefix = Enum('Prefix', prefix_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    
    class Equipped(Enum):
        
        Helm = 0
        Amulet = 1
        Armor = 2
        Greaves = 3
        Armband = 4
        Left_Ring = 5
        Right_Ring = 6
        Right_Hand = 7
        Left_Hand = 8
        Alternate_Right_Hand = 9
        Alternate_Left_Hand = 10
        Artifact = 11

    
    def __init__(self, file_name):
        self.file_name = file_name
        if file_name.split('/')[-1].lower() == 'player.chr':
            self.file_type = 'c'
            with open(self.file_name, 'rb') as handle:
                self.data = bytearray(handle.read())
                self.data_backup = self.data
        else:
            self.file_type = 'd'
            self.xml = ElementTree.parse(self.file_name)
            self.xml_backup = ElementTree.parse(self.file_name)
            self.data = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
            self.data_backup = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
        self.unpack()
        self.inventory_node = self.root['children'][self.Inventory_seg_num]
        self.equip_node = self.root['children'][self.Equip_seg_num]
        self.skill_node = self.root['children'][self.Skill_seg_num]
            
    @staticmethod
    def int2bytes(i, length):
        return i.to_bytes(length, byteorder='little')
    
    @staticmethod
    def bytes2int(b):
        return int.from_bytes(b, byteorder='little')
    
    @staticmethod
    def float2bytes(f):
        return bytearray(struct.pack('<f', f))
    
    @staticmethod
    def bytes2float(b):
        return struct.unpack('<f', b)[0]
        
    @staticmethod
    def find_blocks(data, s, case_sensitive=False):
        blocks = []
        index = 0
        if case_sensitive == True:
            s = s.encode()
            while index < len(data):
                f = data[index:].find(s)
                if f != -1:
                    index += f
                    blocks.append(index)
                    index += 1
                    continue             
                else:
                    break
            return blocks
        elif isinstance(s, str):
            s = s.encode('unicode-escape').decode()
            blocks = [_.start() for _ in re.finditer(s, data.decode('latin-1'), flags=re.I)]
            return blocks
        elif isinstance(s, re.Pattern):
            blocks = [_.start() for _ in re.finditer(s, data.decode('latin-1'))]
            return blocks
            
        
    def split_block(self):
        blocks = ([{'type': 'start', 'index': i} for i in self.find_blocks(self.data, 'begin_block', case_sensitive=True)] + 
                  [{'type': 'end', 'index': i} for i in self.find_blocks(self.data, 'end_block', case_sensitive=True)])
        return sorted(blocks, key=lambda x:x['index'])
    
    def unpack(self, modified=False):
        if modified:
            self.data = bytearray(b''.join([_['data'] for _ in self.root['children']]))
        count = 0
        self.root = {'id': count,
                     'seq': None,
                     'level': 0,
                     'parent': None,
                     'data': None,
                     'children': [],
                     'type': 'fill'}
        current = self.root
        index = 0
        level = 0
        count += 1
        for b in self.split_block():
            fill = b['index'] - index 
            if b['type'] == 'start':
                if fill > 0:
                    current['children'].append({'id': count,
                                                'seq': None,  #'seq' is the sequence# of children node that has data, as defined to have 'begin_block'
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index']],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['children'].append({'id': count,
                                            'seq': len([_ for _ in current['children'] if _['seq'] is not None]),
                                            'level': level + 1,
                                            'parent': current,
                                            'start': b['index'],
                                            'end': None,
                                            'children': [],
                                            'type': 'block'})
                count += 1
                level += 1
                index = b['index']
                current = current['children'][-1]
                last_op = 'start'
            else:
                if last_op == 'end' and fill > 0:
                    current['children'].append({'id': count,
                                                'seq': None,
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index'] + len('end_block')],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['data'] = self.data[current['start']: b['index'] + len('end_block')]
                current.pop('start')
                current.pop('end')
                current = current['parent']
                index = b['index'] + len('end_block')
                level -= 1
                last_op = 'end'             
        if index < len(self.data):
            current['children'].append({'id': count,
                                        'seq': None,
                                        'level': level + 1,
                                        'parent': current,
                                        'data': self.data[index:len(self.data)],
                                        'children': [],
                                        'type': 'fill'},
                                         )
            
                
    def print_tree(self, root=None, space='  '):
        if root is None:
            root = self.root
        if root['children']:
            for c in root['children']:
                self.print_tree(root=c, space=space)
        else:
            print(space * root['level'], root['level'], root['data'])
            
    def get_segs(self, root=None):
        if root is None:
            root = self.root
        segs = []
        def _get_segs(node):
            if node['children']:
                for c in node['children']:
                    _get_segs(c)
            else:
                segs.append(node)
        _get_segs(root)
        yield from segs

                       
    def pack(self, write=False):
        self.data = b''.join([s['data'] for s in self.get_segs()])
        if self.file_type == 'd':
            self.xml.getroot()[0][1].text = base64.b64encode(self.data).decode('latin-1')
            if write:
                self.xml.write(self.file_name) 
        elif self.file_type == 'c':
            with open(self.file_name, 'wb') as handle:
                handle.write(self.data)

    
    def search(self, s, root=None):
        find = list()
        for seg in self.get_segs(root=root):
            find.extend([(seg, i) for i in self.find_blocks(seg['data'], s)])
        return find
    
    def set_int_value(self, s, value, length):
        find = self.search(s)
        if find:
            start = find[0][1] + len(s)
            find[0][0]['data'][start: start+length] = self.int2bytes(value, length)
        else:
            print(f'{s} not found.')
            
    def get_int_value(self, s):
        find = self.search(s)
        if find and s in self.Int_values.keys():
            start = find[0][1] + len(s)
            return self.bytes2int(find[0][0]['data'][start:start+self.Int_values[s][1]])
        else:
            print(f'{s} not found.')          
                        
    def set_temp_values(self):
        find = self.search('temp')
        values = list(zip(*self.Temp_values))[1]
        if find:
            start = find[0][1] + 4
            find[0][0]['data'][start: start+1] = self.int2bytes(values[0], 1)
            for (seg, index), v in zip(find[1:], values[1:]):
                start = index + 4
                seg['data'][start: start+4] = self.float2bytes(v)
        else:
            print('temp not found') 
            
    def get_temp_values(self):
        find = self.search('temp')
        values = []
        if find:
            start = find[0][1] + 4
            values.append(self.bytes2int(find[0][0]['data'][start:start+1]))
            for (seg, index) in find[1:]:
                start = index + 4
                values.append(self.bytes2float(seg['data'][start:start+4]))
            return values
        else:
            print('temp not found') 
            
            
    def get_items(self, itemtype=None, root=None):
        if itemtype is None:
            itemtype = self.Itemtype.Items
        finds = chain(*[self.search('baseName.{4}' + _, root=root) for _ in itemtype.value.values() if pd.notnull(_)])
        return [Item(seg) for (seg, i) in finds]
    
    def map_id(self, df):
        df['_record'] = df['record'].apply(str)
        df = df.merge(self.items_df[['record', 'id']].rename(columns={'record':'_record'}), 
                      left_on='_record', right_on='_record', how='left')
        return df.drop('_record', axis=1)
        
    def equipped_item(self):
        records = [Item(_) for _ in self.get_segs(root=self.root['children'][self.Equip_seg_num]) if 
                re.search('begin_block.*end_block', _['data'].decode('latin-1'), flags=re.DOTALL)]
        df =  pd.DataFrame({'record': records,
                            'Position':self.Equipped,})
        return self.map_id(df)
    
    def inventory_item(self, itemtype=None):
        if itemtype is None:
            itemtype = self.Itemtype.Items
        records = self.get_items(itemtype, root=self.root['children'][self.Inventory_seg_num])
        df = pd.DataFrame({'record': records})
        df['Position'] = df['record'].apply(lambda x: x.get_position())
        return self.map_id(df).sort_values('Position')
    
    def get_skills(self):
        return [Skill(seg) for (seg, i) in self.search(Skill.Pattern, root=self.skill_node)]
    
    @property
    def skills_summary(self):
        return pd.DataFrame([_.summary for _ in self.get_skills()])
    
    def get_skill_insertion_offset(self):
        search = re.search('begin_block.{8}max.{8}', self.skill_node['data'].decode('latin-1'))
        return search.end()
    
    def add_skills(self, skills):
        offset = self.get_skill_insertion_offset()
        current_skills = self.get_skills()
        current_skills_str = [str(_).replace('/', "\\").upper() for _  in current_skills]
        for (skill, level) in skills:
            if hasattr(skill, 'value'):
                skill = skill.value
            skill_str = skill.replace('/', "\\").upper()
            if skill_str in current_skills_str:
                current_skills[current_skills_str.index(skill_str)].assign(Skill.Segs.Level, level)
            else:
                _skill = Skill.make_skill(skill, level)
                self.skill_node['data'][offset:offset] = _skill
                for c in self.skill_node['children'][1:]:
                    c['id'] = c['id'] + 1
                    c['seq'] = None if c['seq'] is None else c['seq']+1
                self.skill_node['children'][1:1] = [{'id': self.skill_node['children'][0]['id']+1,
                                                     'seq': 0,
                                                     'level': self.skill_node['children'][0]['level'],
                                                     'parent': self.skill_node,
                                                     'children': [],
                                                     'data': _skill[:-8],
                                                     'type': 'block'},
                                                    {'id': self.skill_node['children'][0]['id']+2,
                                                     'seq': None,  
                                                     'level': self.skill_node['children'][0]['level'],
                                                     'parent': self.skill_node,
                                                     'data': _skill[-8:],
                                                     'children': [],
                                                     'type': 'fill'}
                                                   ]
    
    @classmethod        
    def validate(cls, record):
        code = cls.Code.get(len(record) - 4, None)
        if code is None:
            return record
        return code.encode() + b'\x00\x00\x00' + record     
     
            
    def optimize(self):
        for item, (value, length) in self.Int_values.items():
            self.set_int_value(item, value, length)
        self.set_temp_values()
        return self
```

```python
class Record:
    
    class Segs(Enum):
        pass
    
    int_values = []
    
    def __init__(self, node):
        self.node = node
        self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))
        
    def __repr__(self):
        return self.match.group(self.Segs.Record.value)
    
    @property
    def summary(self):
        self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))
        return dict([(k, Char.bytes2int(v.encode('latin-1')) if k in self.int_values else v) for k, v in self.match.groupdict().items()])
    
    def assign(self, seg, value, byte_length=2):
        if seg.value in self.int_values:
            value = Char.int2bytes(value, byte_length)
            start = self.match.start(seg.value)
            self.node['data'][start:start+byte_length] = value
        else:
            if hasattr(value, 'value'):
                value = value.value
            start = self.match.start(seg.value) - 4
            end = self.match.end(seg.value)
            self.node['data'][start:end] = Char.validate(value.encode())
            self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))  
```
