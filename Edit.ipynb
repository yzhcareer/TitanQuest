```python
class Char_old:
    
    
    Thunder_blot = b'Records\\Item\\EquipmentWeapon\\Sword\\U_L_ThunderBolt.dbr'
    Three_amazon = b'Records\\Item\\EquipmentWeapon\\Spear\\U_L_FuryoftheThreeAmazons.dbr'
    Theogenes_slaught = b'Records\\Item\\EquipmentWeapon\\Axe\\U_E_Theogenes\'Onslaught.dbr'
    Crystalline_array = b'Records\\Item\\EquipmentWeapon\\Sword\\US_E_CrystallineArray.dbr'
    Geras_stick = b'Records\\Item\\EquipmentWeapon\\Staff\\U_E_Geras\'Walkingstick.dbr'
    Borealis_bow = b'Records\\Item\\EquipmentWeapon\\Bow\\U_E_Borealis.dbr'
    Teucer_bow = b'Records\\Item\\EquipmentWeapon\\Bow\\U_E_Teucer\'sWarbow.dbr'
    Phorkos_trident = b'Records\\Item\\EquipmentWeapon\\Spear\\U_E_Phorkos\'Trident.dbr'
    Athlon = b'Records\\Item\\EquipmentHelm\\U_E_Athlon.dbr'
    Rhesue_armor = b'Records\\Item\\EquipmentArmor\\U_E_Rhesus\'WhitewashedArmor.dbr'
    Secretos_mantle = b'Records\\Item\\EquipmentArmor\\U_E_MantleoftheSecretos.dbr'
    Protean_armor = b'Records\\Item\\EquipmentArmor\\U_E_ProteanArmor.dbr'
    Eidolon_shell = b'Records\\Item\\EquipmentShield\\U_E_EidolonShell.dbr'
    Korybante_shell = b'Records\\Item\\EquipmentShield\\U_E_ShieldoftheKorybantes.dbr'
    Chiron_loop = b'Records\\Item\\EquipmentRing\\U_E_Chiron\'sLoop.dbr'
    Star_stone = b'Records\\Item\\EquipmentRing\\U_E_StarStone.dbr'
    Tartarus_helm = b'Records\\Item\\EquipmentHelm\\US_L_EternalWardofTartarus.dbr'
    Hades_aspis = b'Records\\Item\\EquipmentShield\\U_L_HadesAspis.dbr'
    Pelaran_shield = b'Records\\Item\\EquipmentShield\\U_L_GoldenShieldOfPelaran.dbr'
    Thorny_maul = b'Records\\Item\\EquipmentWeapon\\Club\\U_E_ThornyMaul.dbr'
    Corrupter = b'Records\\Item\\EquipmentWeapon\\Club\\U_L_Saprosthecorrupter.dbr'
    Demeter_sorrow = b'Records\\Item\\EquipmentWeapon\\Club\\U_E_Demeter\'sSorrow.dbr'
    Pagos = b'Records\\Item\\EquipmentWeapon\\Sword\\U_L_Pagos.dbr'
    Pain_weaver = b'Records\\Item\\EquipmentWeapon\\Bow\\U_L_Painweaver.dbr'
    Zhan_lu = b'Records\\Item\\EquipmentWeapon\\Sword\\U_L_Zhanlu.dbr'
    Prometheus_gift = b'Records\\Item\\EquipmentWeapon\\Club\\U_L_Prometheus\'Gift.dbr'
    Apollo_curve_bow = b'Records\\Item\\EquipmentWeapon\\Bow\\U_L_Apollo\'sCurvedBow.dbr'
    Hecate_rod = b'Records\\Item\\EquipmentWeapon\\Staff\\U_L_Hecate\'sRod.dbr'
    Hades_scepter = b'Records\\Item\\EquipmentWeapon\\Staff\\U_L_Hades\'Scepter.dbr'
    Artemis_silver_bow = b'Records\\Item\\EquipmentWeapon\\Bow\\U_L_Artemis\'SilverBow.dbr'
    KuanTi_sorrow = b'Records\\Item\\EquipmentWeapon\\Staff\\U_L_KuanTi\'sSorrow.dbr'
    Orion_mace = b'Records\\Item\\EquipmentWeapon\\Club\\U_L_Orion\'sMace.dbr'
    Nemesis_recurve = b'Records\\Item\\EquipmentWeapon\\Bow\\U_L_Nemesis\'Recurve.dbr'
    Fingerbone_Boreas = b'Records\\Item\\EquipmentWeapon\\Staff\\U_L_FingerboneOfBoreas.dbr'
    Peleus_spear = b'Records\\Item\\EquipmentWeapon\\Spear\\U_L_Peleus\'AshenSpear .dbr'
    Third_paradox = b'Records\\Item\\EquipmentShield\\U_L_Zeno\'sThirdParadox.dbr'
    Ancile = b'Records\\Item\\EquipmentShield\\U_L_Ancile.dbr'
    Athena_mirror = b'Records\\Item\\EquipmentShield\\U_L_AthenasMirroredShield.dbr'
    Priam_gate = b'Records\\Item\\EquipmentShield\\U_L_Priam\'sGate.dbr'
    Archimedes_cogwheel = b'Records\\Item\\EquipmentRing\\U_L_Archimedes\'Cogwheel.dbr'
    Apllo_will = b'Records\\Item\\EquipmentRing\\U_L_Apollo\'sWill.dbr'
    Band_souls = b'Records\\Item\\EquipmentRing\\U_L_BandofSouls.dbr'
    Myrmidon_pendant = b'Records\\Item\\EquipmentAmulet\\U_L_Myrmidon\'sPendant.dbr'
    Immortal_rage = b'Records\\Item\\EquipmentAmulet\\U_L_PendantOfImmortalRage.dbr'
    Aphrodites_favor = b'Records\\Item\\EquipmentAmulet\\U_L_AphroditesFavor.dbr'
    Alexander_armor = b'Records\\Item\\EquipmentArmor\\US_L_Alexander\'sPanoply.dbr'
    BabeiziGi = b'Records\\Item\\EquipmentArmor\\U_L_BabeiziGi.dbr'
    Ward_Tartarus = b'Records\\Item\\EquipmentArmor\\US_L_EternalWardofTartarus.dbr'
    Mantle_sajun = b'Records\\Item\\EquipmentArmor\\U_E_MantleofSajun.dbr'
    Hermes_helm = b'Records\\Item\\EquipmentHelm\\U_L_Hermes\'WingedHelm.dbr'
    Orpheus_garland = b'Records\\Item\\EquipmentHelm\\Um_E_OrpheusGarland.dbr'
    Hesione_veil = b'Records\\Item\\EquipmentHelm\\Um_L_Hesione\'sGoldenVeil.dbr'
    Tracker_helm = b'Records\\Item\\EquipmentHelm\\US_L_Tracker\'sArmor.dbr'
    Alexander_helm = b'Records\\Item\\EquipmentHelm\\US_L_Alexander\'sPanoply.dbr'
    Tracker_greaves = b'Records\\Item\\EquipmentGreaves\\US_L_Tracker\'sArmor.dbr'
    Hermes_greaves = b'Records\\Item\\EquipmentGreaves\\U_L_Hermes\'Talaria.dbr'
    Athena_greaves = b'Records\\Item\\EquipmentGreaves\\U_E_Athena\'sBattleGreaves.dbr'
    Archmage_regalia = b'Records\\Item\\EquipmentGreaves\\USm_L_Archmage\'sRegalia.dbr'
    Alexander_greaves = b'Records\\Item\\EquipmentGreaves\\US_L_Alexander\'sPanoply.dbr'
    Archmage_armband = b'Records\\Item\\EquipmentArmband\\USm_L_Archmage\'sRegalia.dbr'
    
    Eye_of_Ra = b'Records\\xpack\\Item\\Artifacts\\l_da_eyeofra.dbr' #guessed
    Ikon_of_Zeus = b'Records\\xpack\\Item\\Artifacts\\l_da_ikonofzeus.dbr' #guessed
    Thoths_Glory = b'Records\\xpack\\Item\\Artifacts\\l_da_thothsglory.dbr' #guessed
    
    ArchimedsMirror = b'Records\\Item\\Relics\\03_Act1_ArchimedesMirror.dbr'
    Guile = b'Records\\Item\\LootMagicalAffixes\\Suffix\\Default\\Rare_Guile_01.dbr'
    Guild_seed = b'\x80\x13\x00\x00\x04\x00\x00\x00'
    Speed = b'Records\\Item\\LootMagicalAffixes\\Suffix\\Default\\Rare_SpeedAll_03.dbr'
    Speed_seed = b'\xf4r\x00\x00\x04\x00\x00\x00'
    Speed_seed2 = b';I\x00\x00\x04\x00\x00\x00'
    Prism = b'Records\\Item\\LootMagicalAffixes\\Prefix\\Default\\Rare_Prismatic_04.dbr'
    Prism_seed = b'\xac^\x00\x00\x04\x00\x00\x00'  
    Skill2 = b'Records\\Item\\LootMagicalAffixes\\Prefix\\Default\\Rare_SkillAll_02.dbr'
    
    Normal_var = b'\x00\x00\x00\x00\t\x00\x00\x00'
    Relic_var = b'\x03\x00\x00\x00\t\x00\x00\x00'

    
    Code = {30: '\""',
            31: '#',
            32: '$',
            34: '&',
            35: '\'',
            36: '(',
            37: ')',
            38: '*',
            39: '+',
            41: '-',
            42: '.',
            43: '/',
            44: '0',
            45: '1',
            46: '2',
            47: '3',
            48: '4',
            49: '5',
            50: '6',
            51: '7',
            52: '8',
            53: '9',
            54: ':',
            55: ';',
            56: '<',
            57: '=',
            59: '?',
            60: '@'}
    
    Code.update(dict(zip(range(61, 87), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ')))
    
    #Good_seeds = [15, 242, 21547, 24487, 26822, 5557, 11867, 1870, 25954, 13646, 21315, 50, 835, 30234, 5974, 31830]
    
    Inventory_seg_num = 13
    Equip_seg_num = 17
    
    Temp_values = [('Difficulty', 2), 
                   ('Str', 100000),
                   ('Dex', 100000),
                   ('Int', 100000),
                   ('HP', 1000000000),
                   ('MP', 100000)]
    
    Int_values = {'modifierPoints': (60000, 2),
                  'skillPoints': (10000, 2),
                  'money': (1000000000, 4),
                  'experiencePoints': (1000000000, 4),
                  'masteriesAllowed': (2, 1)}
    
    def __init__(self, file_name):
        self.file_name = file_name
        if file_name.split('/')[-1].lower() == 'player.chr':
            self.file_type = 'c'
            with open(self.file_name, 'rb') as handle:
                self.data = bytearray(handle.read())
                self.data_backup = self.data
        else:
            self.file_type = 'd'
            self.xml = ElementTree.parse(self.file_name)
            self.xml_backup = ElementTree.parse(self.file_name)
            self.data = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
            self.data_backup = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
        self.unpack()
        
    @staticmethod
    def int2bytes(i, length):
        return i.to_bytes(length, byteorder='little')
    
    @staticmethod
    def bytes2int(b):
        return int.from_bytes(b, byteorder='little')
    
    @staticmethod
    def float2bytes(f):
        return bytearray(struct.pack('<f', f))
    
    @staticmethod
    def bytes2float(b):
        return struct.unpack('<f', b)[0]
        
    @staticmethod
    def find_blocks(data, s, by_string=True):
        blocks = []
        index = 0
        if by_string == True:
            s = s.encode()
        while index < len(data):
            f = data[index:].find(s)
            if f != -1:
                index += f
                blocks.append(index)
                index += 1
                continue             
            else:
                break
        return blocks
        
    def split_block(self):
        blocks = ([{'type': 'start', 'index': i} for i in self.find_blocks(self.data, 'begin_block')] + 
                  [{'type': 'end', 'index': i} for i in self.find_blocks(self.data, 'end_block')])
        return sorted(blocks, key=lambda x:x['index'])
    
    def unpack(self):
        count = 0
        self.root = {'id': count,
                     'level': 0,
                     'parent': None,
                     'data': None,
                     'children': [],
                     'type': 'fill'}
        current = self.root
        index = 0
        level = 0
        count += 1
        for b in self.split_block():
            fill = b['index'] - index 
            if b['type'] == 'start':
                if fill > 0:
                    current['children'].append({'id': count,
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index']],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['children'].append({'id': count,
                                            'level': level + 1,
                                            'parent': current,
                                            'start': b['index'],
                                            'end': None,
                                            'children': [],
                                            'type': 'block'})
                count += 1
                level += 1
                index = b['index']
                current = current['children'][-1]
                last_op = 'start'
            else:
                if last_op == 'end' and fill > 0:
                    current['children'].append({'id': count,
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index'] + len('end_block')],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['data'] = self.data[current['start']: b['index'] + len('end_block')]
                current.pop('start')
                current.pop('end')
                current = current['parent']
                index = b['index'] + len('end_block')
                level -= 1
                last_op = 'end'             
        if index < len(self.data):
            current['children'].append({'id': count,
                                        'level': level + 1,
                                        'parent': current,
                                        'data': self.data[index:len(self.data)],
                                        'children': [],
                                        'type': 'fill'})
                
    def print_tree(self, root=None, space='  '):
        if root is None:
            root = self.root
        if root['children']:
            for c in root['children']:
                self.print_tree(root=c, space=space)
        else:
            print(space * root['level'], root['level'], root['data'])
            
    def get_segs(self, root=None):
        segs = []
        def _get_segs(node):
            if node['children']:
                for c in node['children']:
                    _get_segs(c)
            else:
                segs.append(node)
        _get_segs(self.root)
        yield from segs
                       
    def pack(self, write=False):
        self.data = b''.join([s['data'] for s in self.get_segs()])
        if self.file_type == 'd':
            self.xml.getroot()[0][1].text = base64.b64encode(self.data).decode('utf-8')
            if write:
                self.xml.write(self.file_name) 
        elif self.file_type == 'c':
            with open(self.file_name, 'wb') as handle:
                handle.write(self.data)
            
    def rollback(self):
        self.xml_backup.write(self.file_name)
    
    def search(self, s, by_string=True):
        find = list()
        for seg in self.get_segs():
            find.extend([(seg, i) for i in self.find_blocks(seg['data'], s, by_string=by_string)])
        return find
    
    def set_int_value(self, s, value, length):
        find = self.search(s)
        if find:
            start = find[0][1] + len(s)
            find[0][0]['data'][start: start+length] = self.int2bytes(value, length)
        else:
            print(f'{s} not found.')
            
    def get_int_values(self, s):
        find = self.search(s)
        if find and s in self.Int_values.keys():
            start = find[0][1] + len(s)
            return self.bytes2int(find[0][0]['data'][start:start+self.Int_values[s][1]])
        else:
            print(f'{s} not found.')          
                        
    def set_temp_values(self):
        find = self.search('temp')
        values = list(zip(*self.Temp_values))[1]
        if find:
            start = find[0][1] + 4
            find[0][0]['data'][start: start+1] = self.int2bytes(values[0], 1)
            for (seg, index), v in zip(find[1:], values[1:]):
                start = index + 4
                seg['data'][start: start+4] = self.float2bytes(v)
        else:
            print('temp not found') 
            
    def get_temp_values(self):
        find = self.search('temp')
        values = []
        if find:
            start = find[0][1] + 4
            values.append(self.bytes2int(find[0][0]['data'][start:start+1]))
            for (seg, index) in find[1:]:
                start = index + 4
                values.append(self.bytes2float(seg['data'][start:start+4]))
            return values
        else:
            print('temp not found')        
            
    @property
    def all_weapons(self):
        find1 = self.search('Records\\Item\\EquipmentWeapon')
        find2 = self.search('Records\\xpack\\Item\\EquipmentWeapon')
        return list(zip(*(find1+find2)))[0]
    
    @property
    def all_shields(self):
        find1 = self.search('Records\\Item\\EquipmentShield')
        find2 = self.search('Records\\xpack\\Item\\EquipmentShield')
        return list(zip(*(find1+find2)))[0]

 
    @property
    def all_rings(self):
        find1 = self.search('Records\\Item\\EquipmentRing')
        find2 = self.search('Records\\xpack\\Item\\EquipmentRing')
        return list(zip(*(find1+find2)))[0]

    
    @property
    def all_amulets(self):
        find1 = self.search('Records\\Item\\EquipmentAmulet')
        find2 = self.search('Records\\xpack\\Item\\EquipmentAmulet')
        return list(zip(*(find1+find2)))[0]


    @property
    def all_armors(self):
        find1 = self.search('Records\\Item\\EquipmentArmor')
        find2 = self.search('Records\\xpack\\Item\\EquipmentArmor')
        return list(zip(*(find1+find2)))[0]


    @property
    def all_helms(self):
        find1 = self.search('Records\\Item\\EquipmentHelm')
        find2 = self.search('Records\\xpack\\Item\\EquipmentHelm')
        return list(zip(*(find1+find2)))[0]

    
    @property
    def all_greaves(self):
        find1 = self.search('Records\\Item\\EquipmentGreaves')
        find2 = self.search('Records\\xpack\\Item\\EquipmentGreaves')
        return list(zip(*(find1+find2)))[0]

    
    @property
    def all_armbands(self):
        find1 = self.search('Records\\Item\\EquipmentArmband')
        find2 = self.search('Records\\xpack\\Item\\EquipmentArmband')
        return list(zip(*(find1+find2)))[0]


    @property
    def all_relics(self):
        find1 = self.search('Records\\Item\\Relics')
        find2 = self.search('Records\\xpack\\Item\\Relics')
        return list(zip(*(find1+find2)))[0]

    
    @property
    def all_charms(self):
        find1 = self.search('Records\\Item\\AnimalRelics')
        find2 = self.search('Records\\xpack\\Item\\AnimalRelics')
        return list(zip(*(find1+find2)))[0]

    
    @property
    def all_artifacts(self):
        find = self.search('Records\\xpack\\Item\\Artifacts')
        return list(zip(*find))[0]
        
    
    @property
    def all_equipments(self):
        return self.all_weapons + self.all_shields + self.all_rings + self.all_amulets + self.all_armors + self.all_helms + self.all_greaves + self.all_armbands + self.all_artifacts
    
    def set_seg_value(self, node, start, end, value_array):
        s = node['data'].find(start.encode()) + len(start.encode())
        e = node['data'].find(end.encode())
        if s >= 0 and e >=0:
            node['data'][s:e] = value_array
        else:
            print('string not found')
            
    def validate(self, record, end_padding=b''):
        code = self.Code.get(len(record) - 4, None)
        if code is None:
            return record
        return code.encode() + b'\x00\x00\x00' + record + end_padding      
            
    def assign_prefix(self, node, value):
        self.set_seg_value(node, 'prefixName', 'suffixName', self.validate(value, end_padding=b'\n\x00\x00\x00'))
        
    def assign_suffix(self, node, value):
        self.set_seg_value(node, 'suffixName', 'relicName', self.validate(value, end_padding=b'\t\x00\x00\x00'))
    
    def assign_relic(self, node, value):
        self.set_seg_value(node, 'relicName', 'relicBonus', self.validate(value, end_padding=b'\n\x00\x00\x00'))
    
    def assign_relicBonus(self, node, value):
        self.set_seg_value(node, 'relicBonus', 'seed', self.validate(value, end_padding=b'\x04\x00\x00\x00'))
        
    def assign_seed(self, node, value):
        if isinstance(value, int):
            value = self.int2bytes(value, 8)
        self.set_seg_value(node, 'seed', 'var1', value)
        
    def assign_var(self, node, value=None):
        if value is None:
            value = self.Normal_var
        self.set_seg_value(node, 'var1', 'end_block', value)
        
    def replace_record(self, record1, record2):
        find = self.search(record1, by_string=False)
        length = len(record1)
        for seg, index in find:
            seg['data'][index - 4 : index + length] = self.validate(record2)   
            
    def optimize(self):
        for item, (value, length) in self.Int_values.items():
            self.set_int_value(item, value, length)
        self.set_temp_values()
        return self
```

```python
class Char:
    
    #Code = (dict(zip(range(30, 87), '"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ')))  
    Code =  {i:chr(i+4) for i in range(30, 123)}
    Inventory_seg_num = 13
    Equip_seg_num = 17
    Skill_seg_num = 3
    Skill_quick_seg_num = 19
    
    Temp_values = [('Difficulty', 2), 
                   ('Str', 100000),
                   ('Dex', 100000),
                   ('Int', 100000),
                   ('HP', 1000000000),
                   ('MP', 100000)]
    
    Int_values = {'modifierPoints': (60000, 2),
                  'skillPoints': (10000, 2),
                  'money': (2000000000, 4),
                  'altMoney': (2000000000, 4),
                  'experiencePoints': (1000000000, 4),
                  'masteriesAllowed': (10, 1),
                  'numberOfDeaths': (0, 1),
                  'max': (None, 2) # length of Char.skills_df
                 }
    
    
    itemtype_df = pd.read_csv("itemclass.csv")
    Itemtype = Enum('Itemtype', itemtype_df.drop('class', axis=1).set_index('category').T.to_dict().items())
    itemtype_pattern = itemtype_df.set_index(['category', 'class']).apply(lambda x: re.compile('|'.join(['(' + re.sub('item(s)?', 'item(s)?', re.escape(_)) + ')' for _ in x.dropna().values]), flags=re.I), axis=1)
    items_df = pd.read_csv("items.csv").dropna(subset=['id'])
    Items = Enum('Items', items_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    skills_df = pd.read_csv("skills.csv")
    Skills = Enum('Skills', skills_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    prefix_df = pd.concat([pd.read_csv(_f) for _f in ["prefix and suffix.csv",
                                                      "prefix.csv"]])
    Prefix = Enum('Prefix', prefix_df[['id', 'record']].set_index('id').to_dict().get('record').items())
    
    Stage_str = re.compile('itemsFoundOverLifetimeRandomizedTotal.{8}temp.{8}hasBeenInGame')
    Stats_str = re.compile('begin_block.{8}temp.{8}temp.{8}temp.{8}temp.{8}temp.{8}end_block')
    
    class Equipped(Enum):
        
        Helm = 0
        Amulet = 1
        Armor = 2
        Right_Ring = 3
        Armband = 4
        Right_Hand = 5
        Left_Ring = 6
        Alternate_Left_Hand = 7
        Artifact = 8
        Greaves = 9
        Alternate_Right_Hand = 10
        Left_Hand = 11

    
    def __init__(self, file_name):
        self.file_name = file_name
        if file_name.split('/')[-1].lower() == 'player.chr':
            self.file_type = 'c'
            with open(self.file_name, 'rb') as handle:
                self.data = bytearray(handle.read())
                self.data_backup = self.data
        else:
            self.file_type = 'd'
            self.xml = ElementTree.parse(self.file_name)
            self.xml_backup = ElementTree.parse(self.file_name)
            self.data = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
            self.data_backup = bytearray(base64.b64decode(self.xml.getroot()[0][1].text))
        self.unpack()
        self.inventory_node = self.root['children'][self.Inventory_seg_num]
        self.equip_node = self.root['children'][self.Equip_seg_num]
        self.skill_node = self.root['children'][self.Skill_seg_num]
        self.skill_quick_node = self.root['children'][self.Skill_quick_seg_num]
            
    @staticmethod
    def int2bytes(i, length):
        return i.to_bytes(length, byteorder='little')
    
    @staticmethod
    def bytes2int(b):
        return int.from_bytes(b, byteorder='little')
    
    @staticmethod
    def float2bytes(f):
        return bytearray(struct.pack('<f', f))
    
    @staticmethod
    def bytes2float(b):
        return struct.unpack('<f', b)[0]
        
    @staticmethod
    def find_blocks(data, s, case_sensitive=False):
        blocks = []
        index = 0
        if case_sensitive == True:
            s = s.encode()
            while index < len(data):
                f = data[index:].find(s)
                if f != -1:
                    index += f
                    blocks.append(index)
                    index += 1
                    continue             
                else:
                    break
            return blocks
        elif isinstance(s, str):
            s = s.encode('unicode-escape').decode()
            blocks = [_.start() for _ in re.finditer(s, data.decode('latin-1'), flags=re.I)]
            return blocks
        elif isinstance(s, re.Pattern):
            blocks = [_.start() for _ in re.finditer(s, data.decode('latin-1'))]
            return blocks
            
        
    def split_block(self):
        blocks = ([{'type': 'start', 'index': i} for i in self.find_blocks(self.data, 'begin_block', case_sensitive=True)] + 
                  [{'type': 'end', 'index': i} for i in self.find_blocks(self.data, 'end_block', case_sensitive=True)])
        return sorted(blocks, key=lambda x:x['index'])
    
    
    def unpack(self, modified=False):
        if modified:
            self.data = bytearray(b''.join([_['data'] for _ in self.root['children']]))
        count = 0
        self.root = {'id': count,
                     'seq': None,
                     'level': 0,
                     'parent': None,
                     'data': None,
                     'children': [],
                     'type': 'fill'}
        current = self.root
        index = 0
        level = 0
        count += 1
        for b in self.split_block():
            fill = b['index'] - index 
            if b['type'] == 'start':
                if fill > 0:
                    current['children'].append({'id': count,
                                                'seq': None,  #'seq' is the sequence# of children node that has data, as defined to have 'begin_block'
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index']],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['children'].append({'id': count,
                                            'seq': len([_ for _ in current['children'] if _['seq'] is not None]),
                                            'level': level + 1,
                                            'parent': current,
                                            'start': b['index'],
                                            'end': None,
                                            'children': [],
                                            'type': 'block'})
                count += 1
                level += 1
                index = b['index']
                current = current['children'][-1]
                last_op = 'start'
            else:
                if last_op == 'end' and fill > 0:
                    current['children'].append({'id': count,
                                                'seq': None,
                                                'level': level + 1,
                                                'parent': current,
                                                'data': self.data[index: b['index'] + len('end_block')],
                                                'children': [],
                                                'type': 'fill'})
                    count += 1
                current['data'] = self.data[current['start']: b['index'] + len('end_block')]
                current.pop('start')
                current.pop('end')
                current = current['parent']
                index = b['index'] + len('end_block')
                level -= 1
                last_op = 'end'             
        if index < len(self.data):
            current['children'].append({'id': count,
                                        'seq': None,
                                        'level': level + 1,
                                        'parent': current,
                                        'data': self.data[index:len(self.data)],
                                        'children': [],
                                        'type': 'fill'},
                                         )
            
                
    def print_tree(self, root=None, space='  '):
        if root is None:
            root = self.root
        if root['children']:
            for c in root['children']:
                self.print_tree(root=c, space=space)
        else:
            print(space * root['level'], root['level'], root['data'])
            
            
    def get_segs(self, root=None):
        if root is None:
            root = self.root
        segs = []
        def _get_segs(node):
            if node['children']:
                for c in node['children']:
                    _get_segs(c)
            else:
                segs.append(node)
        _get_segs(root)
        yield from segs

                       
    def pack(self, write=False):
        self.data = b''.join([s['data'] for s in self.get_segs()])
        if self.file_type == 'd':
            self.xml.getroot()[0][1].text = base64.b64encode(self.data).decode('latin-1')
            if write:
                self.xml.write(self.file_name) 
        elif self.file_type == 'c':
            with open(self.file_name, 'wb') as handle:
                handle.write(self.data)

    
    def search(self, s, root=None):
        find = list()
        for seg in self.get_segs(root=root):
            find.extend([(seg, i) for i in self.find_blocks(seg['data'], s)])
        return find
    

    def set_int_value(self, s, value, length):
        find = self.search(s)
        if find:
            start = find[0][1] + len(s)
            find[0][0]['data'][start: start+length] = self.int2bytes(value, length)
        else:
            print(f'{s} not found.')
            
     
    def get_int_value(self, s):
        find = self.search(s)
        if find and s in self.Int_values.keys():
            start = find[0][1] + len(s)
            return self.bytes2int(find[0][0]['data'][start:start+self.Int_values[s][1]])
        else:
            print(f'{s} not found.')  


    def get_temp_segs(self):
        stage_node = self.search(self.Stage_str)[0][0]
        stats_node = self.search(self.Stats_str)[0][0]
        stage = self.search('temp', root=stage_node)
        stats = self.search('temp', root=stats_node)
        return stage + stats
    
    
    def set_temp_values(self):
        find = self.get_temp_segs()
        values = list(zip(*self.Temp_values))[1]
        if find:
            start = find[0][1] + 4
            find[0][0]['data'][start: start+1] = self.int2bytes(values[0], 1)
            for (seg, index), v in zip(find[1:], values[1:]):
                start = index + 4
                seg['data'][start: start+4] = self.float2bytes(v)
        else:
            print('temp not found') 
        
            
    def get_temp_values(self):
        find = self.get_temp_segs()
        values = []
        if find:
            start = find[0][1] + 4
            values.append(self.bytes2int(find[0][0]['data'][start:start+1]))
            for (seg, index) in find[1:]:
                start = index + 4
                values.append(self.bytes2float(seg['data'][start:start+4]))
            return values
        else:
            print('temp not found') 
            
            
    def get_items(self, itemtype=None, root=None):
        if itemtype is None:
            itemtype = self.Itemtype.Items
        finds = chain(*[self.search('baseName.{4}' + _, root=root) for _ in itemtype.value.values() if pd.notnull(_)])
        return [Item(seg) for (seg, i) in finds]
    
    
    def map_id(self, df):
        df['_record'] = df['record'].apply(str)
        df = df.merge(self.items_df[['record', 'id']].rename(columns={'record':'_record'}), 
                      left_on='_record', right_on='_record', how='left')
        return df.drop('_record', axis=1)
    
        
    def equipped_item(self):
        records = [Item(_) for _ in self.get_segs(root=self.equip_node) if re.search('begin_block.*end_block', _['data'].decode('latin-1'), flags=re.DOTALL)]
        df =  pd.DataFrame({'record': records,
                            'Position':self.Equipped._member_names_})
        return self.map_id(df)
    
    
    def inventory_item(self, itemtype=None):
        if itemtype is None:
            itemtype = self.Itemtype.Items
        records = self.get_items(itemtype, root=self.inventory_node)
        df = pd.DataFrame({'record': records})
        df['Position'] = df['record'].apply(lambda x: x.get_position())
        return self.map_id(df).sort_values('Position')
    
    
    def get_skills(self):
        return [Skill(seg) for (seg, i) in self.search(Skill.Pattern, root=self.skill_node)]
    
    @property
    def skills_summary(self):
        df = pd.DataFrame([_.summary for _ in self.get_skills()])
        df['skillName'] = df['skillName'].apply(lambda x: x.replace('/', '\\'))
        df = pd.concat([df, df['skillName'].apply(lambda x: pd.Series(re.search(r'(skills|quests)\\(?P<Mastery>.*?)\\(?P<id>.*?)\.dbr', x, flags=re.I).groupdict()))]
                       ,axis=1)
        df['Mastery'] = df['Mastery'].str.upper()
        df['id'] = df['id'].apply(lambda x: re.sub('_+', '_', re.sub(r'[^a-zA-Z0-9_]', '_', x.upper())).strip('_'))
        return df.rename(columns={'skillName': 'record'})
    
    
    @classmethod
    def get_item_type(cls, record):
        for (name, pattern) in cls.itemtype_pattern.items():
            if re.search(pattern, record):
                return name
        return None
    
    
    @staticmethod
    def make_id(item):
        if pd.notnull(item['altName']) and not item['category'] in ['Charm', 'Relic', 'Artifact', 'Formula']:
            result = item['altName'].replace(' ', '_').split('_')
        elif re.search(r'^\w{1,3}_\w{1,3}_(set)?\d+$', item['name']):
            result = [item['name'].replace('_', '')]
        elif item['category'] in ['Charm', 'Relic']:
            result = item['name'].split('_')[::-2]
        elif item['category'] == 'Artifact':
            result = item['name'].split('_')[2:]
        elif item['category'] =='Formula':
            result = item['name'].split('_')[2:3]
        elif re.search(r'^\w{1,3}_\w{1,3}_', item['name']):
            result = item['name'].split('_')[2:]
        else:
            result = [item['name']]
        return re.sub(r'_+', '_', re.sub(r'[^a-zA-Z0-9_]', '_', f"{item['category']}_{'_'.join([_.capitalize() for _ in result])}".replace('\'', ''))).strip('_')
        
    
    
    @classmethod
    def make_item_df(cls, items, altNames=dict()):
        df = pd.Series(items).apply(lambda x: pd.Series([x, *Char.get_item_type(x)])).rename(columns={0:'record', 1:'category', 2:'class'})
        df = pd.concat([df, df['record'].apply(lambda x: pd.Series(re.search(r'.*?(?P<pack>xpack\d?)?\\(item|scenery).*\\(?P<name>.*)\.dbr', x, flags=re.I).groupdict())).fillna('Original')], axis=1)
        df['altName'] = df.apply(lambda x: altNames.get(x['record'].lower(), pd.NA), axis=1)
        df['id'] = df.apply(cls.make_id, axis=1)
        df = df.groupby('id').apply(lambda x: x.reset_index())
        df['id'] = df.index.map(lambda x: x[0] if x[1]==0 else f"{x[0]}_{x[1]}")
        df['record'] = df['record'].apply(lambda x: x.lower())
        df = df.set_index('index')   
        return df
    
    
    @property
    def items_summary(self):
        return self.make_item_df([_.summary['record'] for _ in self.get_items()])
    
    
    def get_skill_insertion_offset(self):
        search = re.search('begin_block.{8}max.{8}', self.skill_node['data'].decode('latin-1'))
        return search.end()
    
    
    def add_skills(self, skills):
        offset = self.get_skill_insertion_offset()
        current_skills = self.get_skills()
        current_skills_str = [str(_).replace('/', "\\").upper() for _  in current_skills]
        for (skill, level) in skills:
            if hasattr(skill, 'value'):
                skill = skill.value
            skill_str = skill.replace('/', "\\").upper()
            if skill_str in current_skills_str:
                current_skills[current_skills_str.index(skill_str)].assign(Skill.Segs.Level, level)
            else:
                _skill = Skill.make_skill(skill, level)
                self.skill_node['data'][offset:offset] = _skill
                for c in self.skill_node['children'][1:]:
                    c['id'] = c['id'] + 1
                    c['seq'] = None if c['seq'] is None else c['seq']+1
                self.skill_node['children'][1:1] = [{'id': self.skill_node['children'][0]['id']+1,
                                                     'seq': 0,
                                                     'level': self.skill_node['children'][0]['level'],
                                                     'parent': self.skill_node,
                                                     'children': [],
                                                     'data': _skill[:-8],
                                                     'type': 'block'},
                                                    {'id': self.skill_node['children'][0]['id']+2,
                                                     'seq': None,  
                                                     'level': self.skill_node['children'][0]['level'],
                                                     'parent': self.skill_node,
                                                     'data': _skill[-8:],
                                                     'children': [],
                                                     'type': 'fill'}
                                                   ]
        skill_count = int((len(self.skill_node['children']) - 1) / 2)
        self.set_int_value('max', skill_count, self.Int_values['max'][1])
        
                
    
    @classmethod        
    def validate(cls, record):
        code = cls.Code.get(len(record) - 4, None)
        if code is None:
            return record
        return code.encode() + b'\x00\x00\x00' + record     
     
            
    def optimize(self):
        self.set_temp_values()
        for item, (value, length) in list(self.Int_values.items())[:-1]:  # skill number is automatically set
            self.set_int_value(item, value, length)
        return self
```

```python
class Record:
    
    class Segs(Enum):
        pass
    
    int_values = []
    
    def __init__(self, node):
        self.node = node
        self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))
        
        
    def __repr__(self):
        return self.match.group(self.Segs.Record.value)
    
    @property
    def summary(self):
        self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))
        return dict([(k, Char.bytes2int(v.encode('latin-1')) if k in self.int_values else v) for k, v in self.match.groupdict().items()])
    
    
    def assign(self, seg, value, byte_length=2):
        if seg.value in self.int_values:
            value = Char.int2bytes(value, byte_length)
            start = self.match.start(seg.value)
            self.node['data'][start:start+byte_length] = value
        else:
            if hasattr(value, 'value'):
                value = value.value
            start = self.match.start(seg.value) - 4
            end = self.match.end(seg.value)
            self.node['data'][start:end] = Char.validate(value.encode())
            self.match = re.search(self.Pattern, self.node['data'].decode('latin-1'))  
```

```python
class Skill(Record):
    
    
    All_Skills = Char.skills_df[['record', 'skillLevel']].values.tolist()
    Basic_Skills = All_Skills[:4]
 
    
    class Segs(Enum):
        
        Record = 'skillName'
        Level = 'skillLevel'
        Enabled = 'skillEnabled'
        SubLevel = 'skillSubLevel'
        Active = 'skillActive'
        Transition ='skillTransition'
        
    Pattern = re.compile('begin_block.{8}skillName.{4}(?P<'+Segs.Record.value+'>.*).{4}'
                         +'skillLevel(?P<'+Segs.Level.value+'>.{2}).{6}'
                         +'skillEnabled(?P<'+Segs.Enabled.value+'>.{2}).{6}'
                         +'skillSubLevel(?P<'+Segs.SubLevel.value+'>.{2}).{6}'
                         +'skillActive(?P<'+Segs.Active.value+'>.{2}).{6}'
                         +'skillTransition(?P<'+Segs.Transition.value+'>.{2}).{6}end_block',
                         flags=re.I|re.DOTALL)
    
    TemplateStr = Template('begin_blockÎú\x1d°\t\x00\x00\x00skillName${'
                           +Segs.Record.value
                           +'}\n\x00\x00\x00skillLevel${'
                           +Segs.Level.value
                           +'}\x00\x00\x0c\x00\x00\x00skillEnabled${'
                           +Segs.Enabled.value
                           +'}\x00\x00\r\x00\x00\x00skillSubLevel${'
                           +Segs.SubLevel.value
                           +'}\x00\x00\x0b\x00\x00\x00skillActive${'
                           +Segs.Active.value
                           +'}\x00\x00\x0f\x00\x00\x00skillTransition${'
                           +Segs.Transition.value
                           +'}\x00\x00\t\x00\x00\x00end_blockÞÀ\xadÞ\x0b\x00\x00\x00')
    
    int_values = [Segs.Level.value, Segs.Enabled.value, Segs.SubLevel.value, Segs.Active.value, Segs.Transition.value]
    
    
    @classmethod        
    def make_skill(cls, skill, level, enabled=1, sublevel=0, active=0, transition=0):
        if hasattr(skill, 'value'):
            skill = skill.value
        values = dict(zip([_.value for _ in cls.Segs.__members__.values()],
                          [Char.validate(skill.encode('latin-1')).decode('latin-1'), 
                           *[Char.int2bytes(_, 2).decode('latin-1') for _ in [level, enabled, sublevel, active, transition]]]))
        return bytearray(cls.TemplateStr.safe_substitute(values).encode('latin-1'))
```

```python
class Preference:
    
    "savechr is the correct save files to edit"
     
    def __init__(self, folder, target='savechr'):
        folder = os.path.join(folder, 'SaveData', 'Main')
        self.saves = [_ for _ in os.listdir(folder) if _.startswith('__save')]
        self.backups = [_ for _ in os.listdir(folder) if _.startswith('_backup')]
        self.files = {'savebackup': [os.path.join(folder, _, 'Backup', 'player.chr') for _ in self.saves],
                      'savechr':[os.path.join(folder, _, 'player.chr') for _ in self.saves],
                      'backupbackup':[os.path.join(folder, _, 'Backup', 'player.chr') for _ in self.backups],
                      'backupmap':[os.path.join(folder, _, 'World01.map', 'dict') for _ in self.backups]}
        self.chars = [Char(_) for _ in self.files[target]]
        
        
    def optimize_property(self):
        for c in self.chars:
            c.optimize()
            c.add_skills(Skill.All_Skills)
                
    
    def optimize_items(self, charIndex=0, changeEquip=True):
        npetItem = len(Item.Good_Pet_Item)
        if changeEquip:
            for i, j in zip(self.chars[charIndex].get_items(root=self.chars[charIndex].equip_node), Item.Good_Equip_List):
                if j is not None:
                    i.assign(Item.Segs.Record, j)
        for i in self.chars[charIndex].get_items(root=self.chars[charIndex].equip_node):
            i.make_reflect()
        for i, j in zip(self.chars[charIndex].get_items(root=self.chars[charIndex].inventory_node), cycle(Item.Good_Item_List)):
            i.assign(Item.Segs.Record, j)
            i.make_reflect()
        for i in self.chars[charIndex].get_items(root=self.chars[charIndex].equip_node, itemtype=Char.Itemtype.Amulet) + self.chars[charIndex].get_items(root=self.chars[charIndex].inventory_node, itemtype=Char.Itemtype.Amulet):
            i.make_speed_and_reflect()
        for i in self.chars[charIndex].get_items(root=self.chars[charIndex].equip_node, itemtype=Char.Itemtype.Items) + self.chars[charIndex].get_items(root=self.chars[charIndex].inventory_node, itemtype=Char.Itemtype.Items):
            if 'ring' in i.summary['record']:
                i.make_speed_and_reflect()
            
                
    def copy_to_all_chars(self, sourceIndex=0):
        for k in self.chars:
            for segNum in [Char.Inventory_seg_num, Char.Equip_seg_num, Char.Skill_quick_seg_num, Char.Skill_seg_num]:
                k.root['children'][segNum] = self.chars[sourceIndex].root['children'][segNum]
            k.pack(write=True)  
            
    
    def optimize_all(self, sourceIndex, changeEquip=True):
        self.optimize_property()
        self.optimize_items(charIndex=sourceIndex, changeEquip=changeEquip)
        self.copy_to_all_chars(sourceIndex=sourceIndex)
        for c in self.chars:
            c.pack(write=True)
```
